```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import os
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
os.makedirs("outputs", exist_ok=True)
print("Libraries imported successfully!")
```

```python
df = pd.read_csv("country_wise_latest.csv")

print("Dataset loaded successfully!")
print("Shape of dataset:", df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z ]", "", text)
    return text

df["Country/Region"] = df["Country/Region"].apply(clean_text)
df["WHO Region"] = df["WHO Region"].apply(clean_text)

print("Text columns cleaned successfully!")
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df["Country/Region"] = encoder.fit_transform(df["Country/Region"])
df["WHO Region"] = encoder.fit_transform(df["WHO Region"])

print("Categorical data encoded!")
```

```python
import numpy as np

df.replace([np.inf, -np.inf], np.nan, inplace=True)
print("Infinity values replaced with NaN")
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

print("NaN values filled with mean")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Numerical features scaled between 0 and 1!")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Confirmed"])
plt.title("Distribution of Confirmed COVID-19 Cases")
plt.savefig("outputs/confirmed_cases_distribution.png")
plt.show()
```

```python
df.to_csv("outputs/cleaned_covid_data.csv", index=False)
print("Cleaned COVID-19 dataset saved successfully!")
```
